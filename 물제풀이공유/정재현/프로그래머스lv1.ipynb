```python
def solution(n):
    tempList = []
    
    idx = 0
    while idx < n:
        
        if idx % 2 == 0:
            tempList.append('수')
            idx += 1
        
        else:
            tempList.append('박')
            idx += 1
    return ''.join(tempList)

solution(3)
```
