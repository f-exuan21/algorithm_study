```python
from collections import deque

def solution(progresses, speeds):
    answer = []
    days = 0
    count = 0
    progresses = deque(progresses)
    speeds = deque(speeds)
        
    while progresses:
        if progresses[0] + (days * speeds[0]) >= 100:
            progresses.popleft()
            speeds.popleft()
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            days += 1
    answer.append(count)    
    return answer


n = [93,30,55]
s = [1,30,5]
solution(n, s)

```

```python
def solution(arr):
    answer = []
    min = arr[0]
    index = 0

    for i in range(1, len(arr)):
        if arr[i] < min:
            index = i
            min = arr[i]

    del arr[index]

    if len(arr) == 0:
        answer = [-1]
    else:
        answer = arr

    return answer


arr = [4, 3, 2, 1]
print(solution(arr))
```

```python
def solution(sizes):
    answer = 0

    for i in sizes:
        tmp = 0
        if i[0] < i[1]:
            tmp = i[0]
            i[0] = i[1]
            i[1] = tmp

    max_w = sizes[0][0]
    max_h = sizes[0][1]

    for i in range(1, len(sizes)):
        if sizes[i][0] > max_w:
            max_w = sizes[i][0]

        if sizes[i][1] > max_h:
            max_h = sizes[i][1]

    answer = max_w * max_h

    return answer


sizes = [[60, 50], [30, 70], [60, 30], [80, 40]]
print(solution(sizes))
```

```python
def solution(n, words):
    answer = [0,0]
    said = [words[0]]
    num = 0
    for i in range(len(words)):
        num %= n
        if i >= 1:
            if words[i] in said or words[i-1][-1] != words[i][0]:
                answer = [num + 1, i//n+1]
                break
        num += 1
        said.append(words[i])

    return answer

words = ["tank", "kick", "know", "wheel", "land", "dream", "mother", "robot", "tank"]
solution(3, words)
```
